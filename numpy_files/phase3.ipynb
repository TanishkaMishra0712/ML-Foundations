```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print("===ZOMATO SALES ANLYSIS===")
print("SHAPE:", data.shape)
print("SAMPLE DATA FIRST 3 SALES: \n", data[0:3])
print(data[:3,:])
print(data[:,1:])
```

```python
print("===YEARLY SALE OF EACH RESTAURANT")
print(np.sum(data[:,1:],axis=0))
# print(np.sum(data[:,1:]))    #sums everything in range be it row or column
print("MIN SALE PER RESTAURANT: ", np.min(data[:,1:], axis=1))
print("max sale per year: ", np.max(data[:,1:], axis=0))
print("AGV SALE PER RESTAURANT: ", np.mean(data[:,1:], axis=1))
# print("CUMULATIVE SUM: ", np.cumsum(data[:,1:]))   # NO AXIS GIVES. --> ARRAY FLATTENED AND CUMSUM EXECUTED
cumsum=np.cumsum(data[:,1:], axis = 1)
print("CUMSUM SALE: ", np.cumsum(data[:,1:], axis = 1))
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum, axis=0))
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.title("AVERAGE SALE OF ZOMATO EACH YEAR ACROSS ALL RESTAURANT")
plt.show()
```

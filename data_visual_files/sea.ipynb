Data is of two types:
1) quantative  (quantitavie value without any specific answer like age, salaray)
2) qualitatve  (categorical values such as gender)
and hence we have differnet types of plots for each of them!

1)distplot/histplot
2)pairplot
3)rugplot
4)jointplot

--> for quantative data distibution plots is widely used.


```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
df=sns.load_dataset('tips')
df 
```

WE CAN SEE THAT SEX, SMOKER, DAY, TIME IS CATEGORICAL AND TOTAL BILL AND TIP IS QUANTITAtive but what about size?

```python
df['size'].unique()  # only 6 values are there in size therefore it is also categorical
```

DISTRIBUTION PLOTS:

```python
sns.histplot(df['total_bill'] )
```

```python
sns.histplot(df['total_bill'] , bins=20)
```

```python
sns.histplot(df['total_bill'] , bins=20 , kde=True)  #kernel density estimator(bell shaped curve which tells us the skewness of our data(left or right etc))
```

```python
plt.subplot(1,2,1)      #used to compaer two histograms alongside
sns.histplot(df['total_bill'] , bins=20 , kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'] , bins=20 , kde=True , )
```

JOINTPLOT ( for graph of multiple value)
1) x axis has values from minimun to maximum and same for y axis
2) and then values of column x and y are kept beside each other and pairs are created of x and y values which are then mapped on the axis

```python
sns.jointplot(x='total_bill' , y='tip' , data=df)   #automatically creates scatter plot(kind)
```

```python
sns.jointplot(x='total_bill' , y='tip' , data=df , kind='kde')
```

```python
sns.jointplot(x='total_bill' , y='tip' , data=df , kind='hex')
```

```python
sns.jointplot(x='total_bill' , y='tip' , data=df , kind='reg')
```

PAIRPLOT:

```python
sns.pairplot(df)   # gives mulitple graphs of numerical values( don't confuse with quantative, itis numerical values hence include size as well)
```

```python
sns.pairplot(df , hue='sex')  #displays the data on the basis of male and female
```

```python
sns.pairplot(df , hue='sex' , palette='rainbow')  # now colour is different
```

```python
sns.pairplot(df, hue='size' , palette='rainbow') #more colurs because size has 6-7 types as a catehorical values)
```

RUGPLOT( NOT USED MUCH AND LOOKS LIKE A RUG):

```python
sns.rugplot(df)
```

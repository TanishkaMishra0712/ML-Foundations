MATPLOTLIB:
Basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
x
```

```python
y= x**3
y
```

```python
plt.plot(x,y)
plt.title("FIRST PLOT")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

SUBPLOT IN MATPLOTLIB:

```python
plt.subplot(2,2,1)   # a matrix with 2 row and 2 column is created and first is chosen to plot
plt.plot(x**2,x)     # the plot details are given just after selecting the plot number (the plot number goes like this 1,2,3,4 across 1row then onto anothe)
plt.subplot(2,2,2)
plt.plot(y**2, y)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x**2,y**2)
```

OBJECT ORIENTED WAY

```python
fig=plt.figure()
```

```python
fig=plt.figure()
axis1= fig.add_axes([0.1,0.1,0.8,0.8])   #[distance from left, from right, width, height]
axis1.plot(x,y)
```

```python
fig=plt.figure()      #playing with the axis
axis1= fig.add_axes([0.1,0.1,0.1,0.8])
axis1.plot(x,y)
```

```python
fig=plt.figure()      #playing with the axis
axis1= fig.add_axes([0.1,0.1,0.8,0.1])     
axis1.plot(x,y)
```

```python
fig=plt.figure()      #playing with the axis
axis1= fig.add_axes([0.1,0.7,1,1])  #since the initial size of canvas is not fixed , distance from left and right does not work and behave normally
axis1.plot(x,y)  
```

```python
fig=plt.figure()      #playing with the axis
axis1= fig.add_axes([0.1,0.1,0.1,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.7,0.7,0.3,0.2])
```

```python
fig=plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2= fig.add_axes([0.9,0.2,0.3,0.2])
axis2.plot(x,y**2)
```

FIGSIZE AND DPI( DOTS PER INCH)

```python
fig=plt.figure(figsize=(5,2) , dpi=100)   #100 dpi is standard
axis1=fig.add_axes([0.1,0.1,0.7,0.5])
axis1.plot(x,y)
```

```python
fig=plt.figure(figsize=(10,10) , dpi=100)   #100 dpi is standard
axis1=fig.add_axes([0.1,0.1,0.7,0.5])
axis1.plot(x,y)
```

```python
fig=plt.figure(figsize=(10,10) , dpi=100)   #100 dpi is standard
axis1=fig.add_axes([0.1,0.1,0.7,0.3])
axis1.plot(x,y)
```

```python
fig=plt.figure(figsize=(8,6),dpi = 200 )   # MUCH BRIGHTER AND CLEARER IMAGE
axis1=fig.add_axes([0.1,0.1,0.7,0.7])
axis1.plot(x,y)
```

SAVING IMAGES

```python
fig=plt.figure()
axis= fig.add_axes([0.1,0.1,0.6,0.4])
axis.plot(x,y)
plt.savefig("basicplot.png")
```

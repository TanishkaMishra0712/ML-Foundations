AS SEEN ABOVE ANY CHANGES IN new_df will be reflected on newdf2 as well

COPY: 
ALL THE DATA IS COPIED IN ANOTHER VARIABLE USING 2 METHODS:
1) newdf2=new_df[:]       (shallow copy --> row and coloumn is copied but data cell points to the original dataframe. i.e., changing the data will change original array as well)

2) newdf=new_df.copy      (deep copy --> everything will be copied. therefore, changing data cells here wil not change the original data frame)

```python
newdf3=new_df[:]
newdf3[0][0]="vadakam"     #data will get changed in original data frame as well
new_df.head()
```

```python
newdf2=new_df.copy()
newdf2[0][1]="baingan"      # data will not be changed in the original dataframe
newdf2.head()
```

```python
new_df.head()
```

DROPPING A COLUMN:
1) WE CAN USE DROP() FUNCTION BUT KEEP IN MIND IT DOES NOT DROP THE COLUMN FROM THE ORIGINAL DATAFRAME. IT DREATES A COPY AND THEN DOES IT SO WE HAVE TO ASSIGN IT TO ANOTHER VARIABLE TO EXTRACT THE MODIFIED DATAFRAME.
2) IF WE DONT'T WANT TO REASSIGN THEN WE CAN ALSO USE THE INPLACE = True IN THE ARFUEMENT TO ACTUALLY REPLACE THE VALUE BUT IT IS NOT ADVISED

```python
new_df.drop(0, axis=1)     # actual dataframe is not being changed 
new_df.head()              # therefore actual dataframe is being printed
```

```python
print(new_df.drop(0,axis=1))    # prints the changed dataframe

```

```python
new_df=new_df.drop(0, axis=1)              #  Permanently deletes the col
new_df
```

```python
new_df.loc[0,'F'] = "WALLAH"     #  ADDING ANOTHER COLUMN
new_df
```

```python
new_df.drop('F',axis=1, inplace=True)    #actually drops from the real dataframe as well
new_df
```

PLAYING WITH NULL VALUES:

```python
df= pd.DataFrame({
    "NAME" : ["akash","prem","payal","madhu"],
    "AGE" : [np.nan, np.nan , 19 , 20],
    "DOB": [ pd.NaT , pd.Timestamp(" 25 - 11 - 2003") , pd.Timestamp(" 06- 09 - 2004") , pd.NaT ]})
df
```

```python
df.dropna()     # Default axis=0. Therefore it drops rows with nan
```

```python
df
```

DROPPING COLUMNS WITH NULL VALUES:

```python
df.dropna(axis=1)     # drops col with even a single nan value
```

DROPPING ROWS WITH  ALL VALUE AS NULL:

```python
df.dropna(how="all")      # how="all" --> drops where all value is nan
```

```python
df.dropna(how="any")    # how="any" --> drops where even a single value is nan ( how= any and axis = 0  is the default setting for dropping)
```

DROPPING COLUMNS WITH ALL VALUE AS NULL:

```python
df.dropna(axis = 1 , how="all")   # drops col where all values are nan
```

```python
df.dropna(axis=1, how="any")   # drops the column where even a single value is nan
```

```python
df
```

DROPPING ROWS WHERE ONLY A CERTAIN COLUMN IS nan USING SUBSET:
UNDERSTAND AXIS ( SUM AND DROP SHOW DIFFERENT FUNCTIONALITY. WHY?)

```python
df.dropna(subset= ["AGE" , "DOB" ])    # DROPS ROWS WHERE EITHER AGE OR DOB IS NAN
```
